```python
import pandas as pd
import numpy as np
#from matlplotlib import pyplot as pl

df = pd.read_csv('C:\\Users\\Lenovo\\Downloads\\DCPractice1.csv')
df.head()
```

```python
mean_Ct = df['Consumer Total'].mean()
mean_Cpt = df['Corporate Total'].mean()

df["Consumer Total"].fillna(mean_Ct,inplace=True)
df["Corporate Total"].fillna(mean_Cpt,inplace=True)

con =np.array(df['Consumer Total'],dtype='i')
cop= np.array(df['Corporate Total'],dtype='i')
```

```python
import matplotlib.pyplot as pl
pl.plot(con,cop)
pl.title("Consumer Total vs Corporate Total")
pl.show()
```
